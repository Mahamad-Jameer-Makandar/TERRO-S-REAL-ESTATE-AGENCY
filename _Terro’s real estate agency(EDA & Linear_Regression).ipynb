```python
data['AVG_PRICE'] = np.log10(data.AVG_PRICE)
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
terrodata = scaler.fit_transform(data)
```

```python
data1 = pd.DataFrame(terrodata, columns=data.columns)
```

```python
data1.head()
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression

regression_model = LinearRegression()
```

```python
x = data1.drop(['AVG_PRICE'], axis=1)
```

```python
y = data1[['AVG_PRICE']]
```

```python
x_train,x_test,y_train,y_test = tts(x,y, random_state=1, test_size=0.30)
```

```python
regression_model.fit(x_train,y_train)
```

```python
regression_model.score(x_train, y_train)
```

```python
regression_model.score(x_test, y_test)
```

```python
import statsmodels.formula.api as smf
```

```python
data_train_1 = pd.concat([x_train,y_train], axis=1)
```

```python
data_train_1.head()
```

```python
lm2 = smf.ols(formula= 'AVG_PRICE ~ CRIME_RATE+AGE+INDUS+NOX+DISTANCE+TAX+PTRATIO+AVG_ROOM+LSTAT', data = data_train_1).fit()
print(lm2.params)
```

```python
print(lm2.summary())
```

```python
y_pred = regression_model.predict(x_test)
```

```python
plt.scatter(y_test['AVG_PRICE'], y_pred)
plt.show()
```
